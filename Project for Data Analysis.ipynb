### Data Analysis with Python (Pandas & Matplotlib)
This data came from https://www.kaggle.com/datasets/imtkaggleteam/agriculture-dataset-karnataka

###  #### Data Collection and Loading 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
reviews = pd.read_csv(r"C:\Users\sagar\Documents\Projects\Notebooks\Data project requirement\data_season.csv")
print(reviews.head())
print(reviews.info())
```

### Data Cleaning and Preparation 

```python
print("\nMissing Values:")
print(reviews.isnull().sum())
```

### Data Type Adjustments

```python
reviews['price'] = reviews['price'].apply(lambda x: f"${x}")
print (reviews['price'].head())
```

### Feature Engineering

```python
reviews['weather']= reviews['Temperature'].apply(lambda x: 'High' if x > 40 else 'Low')
print(reviews[['Location', 'weather']].head())
```

###  Exploratory Data Analysis (EDA)

```python
weather_classification = reviews['weather'].value_counts()
print("\nWeather Classification Counts:")
print(weather_classification)
```

### Data Visualizations

```python
plt.figure(figsize=(9, 5))
sns.countplot(data=reviews, x='weather', hue='weather', palette="coolwarm",)
plt.title("Weather Classification Test")
plt.xlabel("Weather")
plt.ylabel("Count")
plt.show()
```

### Optional: Additional visuals

```python
plt.figure(figsize=(9, 5))
sns.histplot(reviews['Temperature'], bins=15, kde=True, color='green')
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
